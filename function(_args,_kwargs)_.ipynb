2. Talabalar haqidagi ma'lumotlarini lug'at ko'rinishida qaytaruvchi funkisya yozing. Talabaning ismi va familiyasi majburiy argument, qolgan ma'lumotlar esa ixtiyoriy ko'rinishda istalgancha berilishi mumkin bo'lsin.

# JAVOBLAR



```python
#Istalgancha sonlarni qabul qilib, ularning ko'paytmasini qaytaruvchi funksiya yozing
def kopaytma(*sonlar) :
  """Istalgancha sonlarni qabul qilib ularning ko'paytmasini qaytaruvchi funksiya"""
  result = 1
  for son in sonlar:
    result = result * son
  return result

kopaytma(1,2,5,8)
```

```python
#Talabalar haqidagi ma'lumotlarini lug'at ko'rinishida qaytaruvchi funkisya yozing. Talabaning ismi va familiyasi majburiy argument, qolgan ma'lumotlar esa ixtiyoriy ko'rinishda istalgancha berilishi mumkin bo'lsin.
def talaba_malumoti(ism, familiya, **malumotlar) :
  """Talabalar haqidagi malumotlarni lug'at haqida qaytaruvchi funksiya"""
  malumotlar['ism'] = ism
  malumotlar['familiya'] = familiya
  return malumotlar

talaba_malumoti('Ibrat', "Ro'ziboyev", yosh = 26, manzil = "Xorazm viloyati")
```
