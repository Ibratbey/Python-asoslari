```python
#Foydanaluvchidan ismi, familiyasi, tug'ilgan yili, tug'ilgan joyi, email manzili va telefon raqamini qabul qilib, lug'at ko'rinishida qaytaruvchi funksiya yozing. Lug'atda foydalanuvchu yoshi ham bo'lsin. Ba'zi argumentlarni kiritishni ixtiyoriy qiling (masalan, tel.raqam, el.manzil)
def mijoz_malumoti(ism, familiya, t_yil, email = '', tel_raqam = ''):
  """Foydalanuvchidan malumotlarni qabul qilib lug'at ko'rinishida qaytaradigan funksiya"""
  info = {
      'ismi' : ism,
      'familiyasi' : familiya,
      'tug\'ilgan yili' : t_yil,
      'email' : email,
      'telefon raqami' : tel_raqam
  }
  return info

foydalanuvchi = mijoz_malumoti('Abdulla', 'Fayziyev', 1995, 'example@gmail.com')
print(foydalanuvchi)
```

```python
#Yuqoridagi funksiyani while yordamida bir necha bor chaqiring, va mijozlar degan ro'yxatni shakllantiring. Ro'yxatdagi mijozlar haqidagi ma'lumotni konsolga chiqaring.
mijozlar = []
while True:
  ism = input("Ismi : ")
  familiya = input("Familiyasi : ")
  t_yil = input("Tug'ilgan yili :")
  email = input("Email :")
  tel_raqam = input("Telefon raqami :")
  mijoz = mijoz_malumoti(ism, familiya, t_yil, email, tel_raqam)
  mijozlar.append(mijoz)
  jarayon = input("Yana mijoz qo'shasizmi? (ha/yo'q)")
  if jarayon != 'ha':
    break

print(mijozlar)
```

```python
#Uchta son qabul qilib, ulardan eng kattasini qaytaruvchi funksiya yozing
def max_top(x, y, z) :
  """ 3 ta son qabul qilibb eng kattasini qaytaruvchi funksiya """
  max = x
  if y >= max :
    max = y
  elif z  >= max :
    max = z

  return max

max_top(4,48,-5)
```

```python
#Foydalanuvchidan aylaning radiusini qabul qilib olib, uning radiusini, diametrini, perimetri va yuzini lug'at ko'rinishida qaytaruvchi funksiya yozing
def aylana_malumotlari(r):
  """ Foydalanuvchidan aylaning radiusini qabul qilib, uning radiusini, diametrini, perimetri va yuzini qaytaruvchi funksiya """
  aylana = {
      'radiusi' : r,
      'diametri' : 2* r,
      'aylana_uzunligi' : 2 * 3.14 * r,
      'aylananing_yuzi' : 3.14 * r**2
  }

  return  aylana

aylana_malumotlari(7.14)
```

```python
#Berilgan oraliqdagi tub sonlar ro'yxatini qaytaruvchi funksiya yozing (tub sonlar —faqat birga va o'ziga qoldiqsiz bo'linuvchi, 1 dan katta musbat sonlar)
def tub_qaytar(boshi, oxiri) :
  """ Berilgan oraliqdagi tub sonlarni qaytaruvchi funksiya """
  sonlar = list(range(boshi, oxiri+1))
  tub_sonlar = []
  for son in sonlar :
    tub = True
    if son == 1 :
      tub = False
    elif son == 2 :
      tub = True
    else :
      for n in range(2, son):
        if son % n == 0 :
          tub = False

    if tub:
      tub_sonlar.append(son)

  return tub_sonlar

tub_qaytar(3,37)
```
