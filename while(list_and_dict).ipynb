# 18-DARS. `WHILE`, RO'YXATLAR VA LUG'ATLAR

While tsikli yordamida ro'yxatlar bilan ishlash.

# AMALIYOT

1. Foydalanuvchidan buyurtma qabul qiluvchi dastur yozing. Mahsulotlar nomini birma-bir qabul qilib, yangi ro'yxatga joylang.
2. e-bozor uchun mahsulotlar va ularning narhlari lug'atini shakllantiruvchi dastur yozing. Foydalanuvchidan lug'atga bir nechta elementlar (mahsulot va uning narhi) kiritishni so'rang.
3. Yuqoridagi ikki dasturni jamlaymiz. Foydalanuvchi buyurtmasi ro'yxatidagi har bir mahsulotni e-bozordagi mahsulotlar bilan solishitiring (tayyor ro'yxat ishlatishingiz mumkin). Agar mahsuot e-bozorda mavjud bo'lsa mahuslot narhini chiqaring, aks holda "Bizda bu mahsulot yo'q" degan xabarni kor'sating.

# JAVOBLAR

```python
buyurtmalar = []
n = 1
while True :
  mahsulot = input(f"{n} - mahsulot nomini kiriting : ")
  buyurtmalar.append(mahsulot)
  n = n + 1
  hohish = input("Yana mahsulot buyurtma qilish uchun 'yes', buyurtmani yakunlash uchun 'no' kiriting : ")
  if hohish.lower() == 'no' :
    break
```

```python
mahsulot_naxlari = {}
ishora = True
print("e-bozor uchun mahsulot va mahsulot narxlarini kiriting!")
while ishora:
  mahsulot = input("Mahsulot nomini kiriting : ")
  narx  = int(input(f"{mahsulot} narxini kiriting : "))
  mahsulot_naxlari[mahsulot] = narx
  javob = input("Yana mahsulot kiritish uchun 'y', to'xtash uchun 'no' kiriting! Yana mahsulot qo'shasizmi? ")
  if javob == 'no' :
    ishora = False

print(mahsulot_naxlari)
```

```python
for buyurtma in buyurtmalar :
  if buyurtma in mahsulot_naxlari :
    print(f"{buyurtma} narxi : {mahsulot_naxlari[buyurtma]} so'm")
  else :
    print(f"Bizda {mahsulot} yo'q")
```
