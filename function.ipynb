3. Foydalanuvchidan son olib, son juft yoki toqligini konsolga chiqaruvchi funksiya yozing.
4. Foydalanuvchidan ikkita son olib, ulardan kattasini konsolga chiqaruvchi funksiya yozing. Agar sonlar teng bo'lsa "Sonlar teng" degan xabarni chiqaring.
5. Foydalanuvchidan x va y sonlarini olib, ni konsolga chiqaruvchi funksiya yozing.

7. Foydalanuvchidan son qabul qilib, sonni 2 dan 10 gacha bo'lgan sonlarga qoldiqsiz bo'linishini tekshiruvchi funksiya yozing. Natijalarni konsolga chiqaring.

# JAVOBLAR

```python
#Foydalanuvchi ismi va yoshini so'rab, uning tug'ilgan yilini hisoblaydigan funksiya yozing.
def t_yil_hisobla(ism, yosh):
  """ Foydalanuvchi ismi va yoshini so'rab, uning tug'ilgan yilini hisoblaydigan funksiya """
  print(f"{ism} siz {2025 - yosh} - yilda tug'ilgan ekansiz!")

t_yil_hisobla("Asad",26)
```

```python
#Foydalanuvchidan son olib, uning kvadrati va kubini konsolga chiqaruvchi funksiya yozing.
def daraja_hisobla(son) :
  """ Foydalanuvchidan son olib, uning kvadrati va kubini konsolga chiqaruvchi funksiya """
  print(f"{son} ning kvadrati : {son ** 2} \n{son} ning kubi esa {son ** 3} ga teng")

daraja_hisobla(5)
```

```python
#Foydalanuvchidan son olib, son juft yoki toqligini konsolga chiqaruvchi funksiya yozing.
def toqmi(son):
  """ Foydalanuvchidan son olib, son juft yoki toqligini konsolga chiqaruvchi funksiya """
  if son % 2 == 0 :
    natija = "juft"
  else :
    natija = "toq"
  print(f"{son} soni {natija} son")

toqmi(97)
```

```python
#Foydalanuvchidan ikkita son olib, ulardan kattasini konsolga chiqaruvchi funksiya yozing. Agar sonlar teng bo'lsa "Sonlar teng" degan xabarni chiqaring.
def taqqosla(son1, son2):
  """ 2 ta sonni taqqoslab kattasini chiqaruvchi funksiya """
  if son1 == son2 :
    print(f"{son1} = {son2}")
  elif son1 > son2 :
    print(f"{son1} > {son2}")
  else :
    print(f"{son2} > {son1}")

taqqosla(67,54)
```

```python
# Foydalanuvchidan x va y sonlarini olib, x^y ni konsolga chiqaruvchi funksiya yozing.
# Yuqoridagi funksiyada y uchun 2 standart qiymatini bering.
def darajaTop(x,y=2):
  """x ning y - darajasini hisoblovchi funksiya """
  print(f"{x} ning {y}-darajasi {x**y} ga teng")

darajaTop(11,3)
darajaTop(7,6)
darajaTop(15)
```
