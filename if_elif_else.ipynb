```python
mahsulotlar = ["olma", "anor", "non", "suv","snikers", "bodom","shakar","salfetka","pecheniya","energetik"]
savat = []
print("5 ta mahsulot kiriting")
for n in range(5):
  savat.append(input(f"{n+1} - mahsulot : "))

bor_mahsulotlar = []
mavjud_emas = []
for mahsulot in savat:
  if mahsulot in mahsulotlar:
    bor_mahsulotlar.append(mahsulot)
  else :
    mavjud_emas.append(mahsulot)
if mavjud_emas:
  print(f"Quyidagi mahsulotlar do'konimizda yo'q : \n{mavjud_emas}")
else:
  print("Siz so'ragan barcha mahsulotlar do'konimizda bo")
```
